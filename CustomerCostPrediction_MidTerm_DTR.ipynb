I am considering a threshold between 0.75 and -0.75 for the feature selection. So all the features between above 0.75 are positively correlated. For example, salad_bar and prepared_food are highly correlated as they have a value of 1. So we can eliminate one of these features. Similarly we can eliminate store_sales(in million), store_cost(in millions), store_sqft, gross_weight, avg_cars_at home(approx).1, meat_sqft.

```python
df  = df.drop(columns=['salad_bar', 'gross_weight', 'avg_cars_at home(approx).1', 'meat_sqft', 'store_sales(in millions)', 'store_cost(in millions)', 'store_sqft'])
df.head()
```

# Training the model on single split

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
```

```python
X = OrdinalEncoder().fit_transform(df.drop(columns='cost')) # Ordinal encoder
y = df['cost']
```

**Creating training and validation set**

```python
train_X, val_X, train_y, val_y = train_test_split(X,y,test_size=0.3) # Training data and Test data split ratio -> 70:30
```

```python
print("Shape of Training X :",train_X.shape)
print("Shape of Validation X :",val_X.shape)
```

```python
print("Shape of Training y :",train_y.shape)
print("Shape of Validation y :",val_y.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import  metrics
```

**Model Building**

```python
dc = DecisionTreeRegressor(random_state = 1)
dc.fit(train_X,train_y)
print("Fitting Decision Tree Regressor")
```

# Results of single split

```python
%%time
pred_val_y = dc.predict(val_X)
print('MAE on Validation set :',metrics.mean_absolute_error(val_y, pred_val_y))
print('MSE on Validation set :',metrics.mean_squared_error(val_y, pred_val_y))
print('RMSE on Validation set :',np.sqrt(metrics.mean_squared_error(val_y, pred_val_y)))
print('R2 Score on Validation set :',metrics.r2_score(val_y, pred_val_y))
```

**Visualizing the Actual values against Predicted Values for single split:**

```python
plt.figure(figsize=(10,10))
plt.scatter(val_y, pred_val_y, c='orange')

p1 = max(max(pred_val_y), max(val_y))
p2 = min(min(pred_val_y), min(val_y))
plt.plot([p1, p2], [p1, p2], 'blue')
plt.xlabel('Actual Values', fontsize=15)
plt.ylabel('Prediction Values', fontsize=15)
plt.legend(['Actual cost value', 'Predicted cost value'], loc='upper left')
plt.axis('equal')
plt.show()
```

# Cross Validation

```python
%%time
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold

cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
mae_cv_scores = cross_val_score(dc, X, y, cv=cv, scoring='neg_mean_absolute_error') # cross validation for mae metric
mse_cv_scores = cross_val_score(dc, X, y, cv=cv, scoring='neg_mean_squared_error') # cross validation for mse metric
rmse_cv_scores = cross_val_score(dc, X, y, cv=cv, scoring='neg_root_mean_squared_error') # cross validation for rmse metric
r2_cv_scores = cross_val_score(dc, X, y, cv=cv, scoring='r2') # cross validation for r2 metric
```

**Cross Validation Results:**

```python
print('MAE scores after cross validation: \n', mae_cv_scores)
print('\nMSE scores afte cross validation: \n', mse_cv_scores)
print('\nRMSE scores after cross validation: \n', rmse_cv_scores)
print('\nR2 scores after cross validation: \n', r2_cv_scores)
```

# Error Bar Plots for all metrics with Mean and Standard Deviation

```python
from numpy import absolute

scores_rmse = absolute(rmse_cv_scores)
scores_mse = absolute(mse_cv_scores)
scores_r2 = absolute(r2_cv_scores)
scores_mae = absolute(mae_cv_scores)
labels = ['RMSE', 'R2', 'MAE']
x_pos = np.arange(len(labels))
CTEs = [absolute(scores_rmse.mean()), absolute(scores_r2.mean()), absolute(scores_mae.mean())]
error = [absolute(scores_rmse.std()), absolute(scores_r2.std()), absolute(scores_mae.std())]
fig, ax = plt.subplots()
ax.bar(x_pos, CTEs,
       yerr=error,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=2)
ax.set_ylabel('Error Bar for Decision Tree Regressor')
ax.set_xticks(x_pos)
ax.set_xticklabels(labels)
ax.set_title('Error Bar plots of RMSE, R2 Score, and MAE')
ax.yaxis.grid(True)

#Print metric values
print('RMSE score: %.8f (SD: %.8f)' % (scores_rmse.mean(), scores_rmse.std()) )
print('R2 score  : %.8f (SD: %.8f)' % (scores_r2.mean(), scores_r2.std()) )
print('MAE score : %.8f (SD: %.8f)' % (scores_mae.mean(), scores_mae.std()) )
print('MSE score : %.7f (SD: %.8f)' % (scores_mse.mean(), scores_mse.std()) )

# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
